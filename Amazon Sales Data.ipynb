### Correlation Analysis:

```python
# Correlation matrix
correlation_matrix = df.corr()
correlation_matrix
```

### Correlation
Correlation tells about the relationship between two variables i.e. two variables are linearly related.

```python
# Correlation between Total Revenue and Total Profit
revenue_profit_correlation = df['Total Revenue'].corr(df['Total Profit'])
print("Correlation between Total Revenue and Total Profit:", revenue_profit_correlation)
```

```python
# Correlation between Uint sold and Total Profit
unit_sold_profit_correlation = df['Units Sold'].corr(df['Total Profit'])
print("Correlation between Unit Sold and Total Profit:", unit_sold_profit_correlation)
```

### Segmentation Analysis:

```python
# Total profit for all Region
plt.figure(figsize=(8,3))
profit_region=df.groupby(['Region'],as_index=False)['Total Profit'].sum().sort_values(by='Total Profit',ascending=False)
sns.barplot(x='Region',y='Total Profit',data=profit_region)
plt.title("Total Profit for each Region",fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar plot of Total Profit by Region
sns.barplot(x='Region', y='Total Profit', data=df)
plt.title('Total Profit by Region')
plt.xticks(rotation=90)
plt.show()
```

```python
# Segment the data by Region and calculate mean Total Profit for each region
profit_by_region = df.groupby(['Region'],as_index=False)['Total Profit'].mean().sort_values(by='Total Profit',ascending=False)
#a=profit_by_region.index
#b=profit_by_region.values
plt.figure(figsize=(9,4))
sns.barplot(x='Region',y='Total Profit',data=profit_by_region)
#sns.barplot(x=a,y=b)
plt.xticks(rotation=90)
plt.title("Average Total Profit of each Region",fontweight='bold')
plt.ylabel('Average Total Profit')
plt.show()
```

```python
region_unit=df.groupby(['Region'],as_index=False)['Units Sold'].sum().sort_values(by='Units Sold',ascending=False)
sns.barplot(x='Region',y='Units Sold',data=region_unit)
sns.set({'figure.figsize':(9,4)})
plt.xticks(rotation=90)
plt.title("Units Sold over Region",fontweight='bold')
plt.show()
```

```python
# Item Type vs Total Units sold
units_sold_by_item=df.groupby('Item Type')['Units Sold'].sum()
a=units_sold_by_item.index
b=units_sold_by_item.values
plt.figure(figsize=(5,5))
plt.pie(b,labels=a,autopct='%.f%%')
plt.title("Total % Units of Item Type sold",fontweight='bold')
plt.show()
```

```python
# Item Type vs Average Revenue
avg_revenue_by_item=df.groupby('Item Type')['Total Revenue'].mean()
a=avg_revenue_by_item.index
b=avg_revenue_by_item.values
plt.figure(figsize=(6,6))
plt.pie(b,labels=a,autopct='%.1f%%')
plt.title("Average Revenue of Item Type",fontweight='bold')
plt.show()
```

```python
sales_chnnel_profit=df.groupby(['Sales Channel'],as_index=False)['Total Profit'].mean().sort_values(by='Total Profit',ascending=False)
plt.figure(figsize=(6,3))
sns.barplot(x='Sales Channel',y='Total Profit',data=sales_chnnel_profit)
plt.xticks(rotation=90)
plt.title("Average Profit by Sales Channel",fontweight='bold')
plt.show()
```

```python
# Box plot of Total Profit by Order Priority
plt.figure(figsize=(5, 5))
sns.boxplot(x='Order Priority', y='Total Profit', data=df)
plt.title('Total Profit by Order Priority',fontweight='bold')
plt.show()
```

##### Note
In above box plot the Order Priority 'C' have most outlier i.e. Total Profit of some Order Priority 'C' is very high.

```python
# Country vs Total Units Sold
plt.figure(figsize=(12,4))
country_unit=df.groupby(['Country'],as_index=False)['Units Sold'].sum().sort_values(by='Units Sold',ascending=False).head(10)
ax=sns.barplot(x='Country',y='Units Sold',data=country_unit)
plt.xticks(rotation=90)
plt.title("Total Units Sold in Top 10 Country",fontweight='bold')
# Add text inside the bars
def add_text(x, **kwargs):
    for i, v in enumerate(x):
        ax.text(i, v + 0.5, str(v), ha='center', va='bottom')

add_text(country_unit['Units Sold'])
plt.show()
```

### Time Series Analysis:

```python
# Convert Order Date and Ship Date to datetime objects
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Extract year and month from Order Date
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

# Total Revenue over month
revenue_by_month = df.groupby(['Order Year', 'Order Month'])['Total Revenue'].sum()
print(revenue_by_month)
```

```python
month = df.groupby(['Order Month'])['Total Revenue'].sum()
```

```python
sns.lineplot(data=month, marker='o', color='r')
plt.title('Total Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue',fontweight='bold')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Total Revenue over year
revenue_by_year = df.groupby(['Order Year'])['Total Revenue'].sum()
print(revenue_by_year)
c=revenue_by_year.index
d=revenue_by_year.values
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x=c,y=d)
plt.title("Total Revenue Over the Year",fontweight='bold')
plt.show()
```

```python
values=revenue_by_month.values
d=list(values)
l=len(d)

# List of year 
a=[]
for i in range(l):
    a.append(revenue_by_month.index[i][0])

# List of months    
b=[]
for i in range(l):
    b.append(revenue_by_month.index[i][1])
    
#Join the both list as combined data as month and year
c= [f'{x}-{y}' for x, y in zip(b,a)]
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x=c,y=d)
plt.title("Total Revenue Over the Months Period",fontweight='bold')
plt.xticks(rotation=90,fontweight='bold')
plt.show()
```

```python
# Histogram of Total Profit
plt.figure(figsize=(8, 6))
sns.histplot(df['Total Profit'], bins=20, kde=True)
plt.title('Distribution of Total Profit',fontweight='bold')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line plot of Total Revenue over time
plt.figure(figsize=(16, 7))
plt.title('Total Revenue over Time')
sns.lineplot(x='Order Date', y='Total Revenue', data=df)
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')
plt.show()
```

```python
d=[]
for i in range(100):
    q=(df['Ship Date'][i]-df['Order Date'][i]).days
    d.append(q)
df['date_diff']=d
```

```python
avg_days_item_type=df.groupby('Item Type')['date_diff'].mean()
a=avg_days_item_type.index
b=avg_days_item_type.values
plt.figure(figsize=(10,6))
sns.barplot(x=a,y=b)
plt.title('Average days to ship order',fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```
